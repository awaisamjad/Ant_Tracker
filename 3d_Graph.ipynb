```python
df_humid = pd.read_csv("Humidity_Values.csv")
```

```python
df_humid.rename(columns={"#!,Workbook:,Custom,C" : "Soil Moisture", "Unnamed: 3" : "Temperature","Unnamed: 5" : "Humidity" }, inplace=True)
```

```python
df_humid.drop(columns={"ustom", "Unnamed: 2", "Unnamed: 4"}, inplace= True)
```

```python
df_humid
```

```python
df_humid.drop([0,1], inplace= True)
```

```python
df_humid["Soil Moisture"] = df_humid["Soil Moisture"].str.replace("Soil Moisture: ", "")
```

```python
df_humid["Soil Moisture"] = pd.to_numeric(df_humid["Soil Moisture"])
```

```python
df_humid["Soil Moisture"]
```

```python
df_humid["Temperature"] = df_humid["Temperature"].str.replace("Temperature: ", "")
```

```python
df_humid["Temperature"] = df_humid["Temperature"].str.replace("c", "")
```

```python
df_humid["Temperature"] = pd.to_numeric(df_humid["Temperature"])
```

```python
df_humid["Humidity"] = df_humid["Humidity"].str.replace("Humidity: ", "")
```

```python
df_humid["Humidity"] = df_humid["Humidity"].str.replace(" %", "")
```

```python
df_humid["Humidity"] = pd.to_numeric(df_humid["Humidity"])
```

```python
df_humid
```

```python
df_humid.iloc[0, df_humid.columns.get_loc('Humidity')] = df_humid["Humidity"].mean()
```
